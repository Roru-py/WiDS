```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
```

```python
movie = pd.read_csv("movies.dat", delimiter="::", engine="python",header = None, names = ['movieId', 'title', 'genre'],encoding="latin-1")
rating = pd.read_csv("ratings.dat", delimiter="::", engine="python",header = None,names=['userId', 'movieId', 'rating', 'timestamp'],encoding = "latin-1")
movie.head()
```

```python
rating.head()
```

```python
movie_rating = pd.merge(movie,rating, on = "movieId")
movie_rating[["movie_name","year"]]= movie_rating["title"].str.split("(",expand=True)[[0,1]]
movie_rating
```

```python
movies = movie_rating.pivot_table(index='userId',columns='movie_name',values='rating',fill_value=0)
movies
```

```python
user = 100
top = 10
```

```python
similarity_users = cosine_similarity(movies)
similar_users = pd.DataFrame(similarity_users,index=movies.index,columns=movies.index)
similar_users 
```

```python
def user_based(user,top):
    sim = similar_users[user].drop(user)
    weights = sim/sim.sum()
    not_watched = movies.loc[movies.index!= user,movies.loc[user,:]==0]
    sim_weights = pd.DataFrame(not_watched.T.dot(weights.to_numpy()),columns=["weighted_avgs"])
    return(sim_weights.sort_values(by="weighted_avgs",ascending=False).head(top))
```

```python
user_based(100,10)
```

```python
def item_based(user,top):
    watched_ratings = movies.loc[movies.index==user,movies.loc[user,:]>0].T
    sim_w_to_nw = similar_movies.loc[movies_new.loc[movies_new.loc[:,user]==0,movies_new.columns == user].index,movies_new.loc[movies_new.loc[:,user]>0,movies_new.columns==user].index]
    df = pd.DataFrame(sim_w_to_nw.dot(watched_ratings.to_numpy()))
    df = df.rename(columns={0:'Pred_rating'})
    den = sim_w_to_nw.T.sum()
    pred = {}
    for i in df.index:
        pred[i] = df.loc[i]/den[i]
    pred_new = pd.DataFrame(pred).T
    return pred_new.sort_values(by='Pred_rating',ascending=False).head(top)
```

```python
def user_eval(user):
    sim = similar_users[user].drop(user)
    weights = sim/sim.sum()
    watched_ratings = movies.loc[movies.index==user,movies.loc[user,:]>0].T
    not_watched_test = watched_ratings.iloc[round(len(watched_ratings.index)*train):]
    not_watched = movies.loc[movies.index!= user,not_watched_test.index]
    sim_weights = pd.DataFrame(not_watched.T.dot(weights.to_numpy()),columns=["weighted_avgs"])
    RMSE = np.sqrt(np.mean(np.square(not_watched_test[user]-sim_weights["weighted_avgs"])))
    return RMSE
```
